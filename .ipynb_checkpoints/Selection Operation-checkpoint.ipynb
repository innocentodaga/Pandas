* this means some columns, rows, or specific records

```python
# Selecting multiple columns i.e. Name and Age hence they are case sensitive
df[["Name", "Age"]]
```

### selecting a specified part of the dataset with rows and columns

using loc and iloc

```python
# Query selecting the first row and second and the fifth columns

df.iloc[0, [1,4]]
```

```python
df.loc[0, ["Survived","Sex"]]
```

```python
# Selecting the first five rows and first three columns

df.iloc[:5, :3]
```

```python
# Querying the first and last row and last three columns
df.iloc[[0, -1], -3:]
```

```python
# Querying the rows index between 5 and 10 and columns from Name to Ticket
df.loc[5:10, "Name":"Ticket"]
```

```python
# Querying a boolean attribute for all male passengers
df.Sex == 'male'
```

```python
# Querying all male passengers from the dataset
male = df.loc[df.Sex == 'male']
male
```

```python
len(male)
```

```python
# selecting all female passengers 
df.loc[df.Sex == "female"]
```

```python
# Query: Select all female passengers who are oveer 35 years

df.loc[(df.Sex == "female")&(df.Age > 35)]
```
