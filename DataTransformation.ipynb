## Data Transformation

* converting data to categorical datatype

```python
import pandas as pd

df = pd.read_csv("datasets/employees.csv")
```

```python
df
```

```python
# To find the frequency of the data i.e of the column gender
df.Gender.value_counts()
```

```python
# to find the number of non values
df.Gender.isna().sum()
```

```python
# CONVERTING TO CATEGORICAL TYPE
pd.Categorical(df.Gender)
```

```python
# Creating a new column
df["Gender2"] = pd.Categorical(df.Gender)
```

```python
df.head()
```

```python
df.memory_usage()
```

```python
import random
random.choices(["Very Low","Low","Medium","High","Very High"], k = 1000)
```

```python
df["Score"] = pd.Series(random.choices(["Very Low","Low","Medium","High","Very High"], k = 1000))
```

```python
df.head()
```

```python
df.dtypes
```

```python
pd.Categorical(df.Score)
```

```python
# Chaging the lexical order of the categorical data
pd.Categorical(df.Score, categories=['High', 'Low', 'Medium', 'Very High', 'Very Low'], ordered=True)
```

```python
# adding the above to the dataframe
df["Score2"] = pd.Categorical(df.Score, categories=['High', 'Low', 'Medium', 'Very High', 'Very Low'], ordered=True)
```

```python
df
```

```python
# to find the  number of people whose score2 > medium of the score2 columns
sum(df.Score2 > 'Medium')
```

```python
# Finding max and min of the salary
df.Salary.min()
```

```python
pd.cut(df.Salary, bins = [35000, 50000,75000,100000,125000, 150000], labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High'])
```

```python
# creating a category column for the above 
df["Salary_Cat"] = pd.cut(df.Salary, bins = [35000, 50000,75000,100000,125000, 150000], labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High'])
```
