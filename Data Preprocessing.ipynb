## Data preprocessing

* Data cleaning and data transformation
* Handling missing values and Duplicates

```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/titanic.csv")
```

```python
df
```

```python
# checking whether our dataset contains missing values or not
df.isna() # this returns a boolean value for the results
```

```python
# to check which column has the empty values 
df.isna().any() # returns a boolean value for the columns if the contain missing values or not
```

```python
# Removing missing values involes removing either from the rows or columns 
# this returns a new dataframe
df2 = df.dropna(axis=0) # 0 removes the the rows with missing values while 1 form columns
df2
```

```python
df2.isna().any()
```

```python
df.shape, df2.shape
```

```python
# to apply the changes made to the original dataframe we set inplace = True
df.dropna(inplace=True)
```

```python
# confriming the shape of the dataframe
df.shape
```

```python
# we have to re-read the dataset to ignore the changes made earlier
df = pd.read_csv("datasets/titanic.csv")
df
```

```python
# Replacing the missing values with certain values
df2 = df.fillna(0) # replacing missing values with 0
df2
```

```python
# filling the missing values with average of the columns with only numerics
df3 = df.fillna(df.mean(numeric_only=True))
```

```python
df3
```

```python
# filling in values for none numeric values
# We use mode to change the missing value with the most frequent item
df4 = df3.fillna(df.mode().iloc[0]) # we set iloc[0] to get the first most frequent item 
```

```python
df4
```

```python
df4.isna()
```

```python
# How to handle duplicate values
# To find duplicate row we use
df.duplicated()
```

```python
# to find duplicates in both rows and columns
df.duplicated().any()
```

```python
# Copying the first three rows and pasting to the end of the dataframe
df_duplicate = pd.concat((df, df.iloc[:3]))
```
